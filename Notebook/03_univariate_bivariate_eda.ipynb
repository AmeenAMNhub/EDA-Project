```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv(r"C:\Users\sainudheen\OneDrive - KEO International Consultants\Desktop\EDA_Project\Data\cleaned_data_day2.csv")
df.head()
```

```python
plt.figure(figsize=(10,6))

sns.boxplot(x='customer_age_group', y='total_spend',hue='customer_age_group', data=df,palette="Set2")

plt.title("Spending Distribution by Age Group", fontsize=14, fontweight="bold")
plt.xlabel("Customer Age Group")
plt.ylabel("Total Spend")
plt.savefig(r"C:\Users\sainudheen\OneDrive - KEO International Consultants\Desktop\EDA_Project\Reports\Figures\Spending Distribution by Age Group", dpi=300, bbox_inches="tight")

plt.show()
```

```python
df['order_hour'] = pd.to_datetime(df['order_time'], format='%H:%M:%S').dt.hour

plt.figure(figsize=(10,6))

sns.countplot(x='order_hour',hue='order_hour', data=df,palette="coolwarm")

plt.title("Order Distribution by Hour of Day", fontsize=14, fontweight="bold")
plt.xlabel("Hour of Day")
plt.ylabel("Number of Orders")
plt.savefig(r"C:\Users\sainudheen\OneDrive - KEO International Consultants\Desktop\EDA_Project\Reports\Figures\Order Distribution by Hour of Day", dpi=300, bbox_inches="tight")

plt.show()
```

```python
plt.figure(figsize=(10,6))

sns.countplot(x='customer_satisfaction',hue='customer_satisfaction', data=df, palette="viridis")

plt.title("Customer Satisfaction Ratings", fontsize=14, fontweight="bold")
plt.xlabel("Customer Satisfaction Rating")
plt.ylabel("Number of Customers")

plt.savefig(r"C:\Users\sainudheen\OneDrive - KEO International Consultants\Desktop\EDA_Project\Reports\Figures\Customer Satisfaction Ratings", dpi=300, bbox_inches="tight")
plt.show()
```

```python
num_cols = df.select_dtypes(include=['int64','float64'])

plt.figure(figsize=(10,6))

sns.heatmap(num_cols.corr(),
            annot=True,
            cmap="coolwarm",
            fmt=".2f")

plt.title("Correlation Between Numerical Features", fontsize=14, fontweight="bold")

plt.savefig(r"C:\Users\sainudheen\OneDrive - KEO International Consultants\Desktop\EDA_Project\Reports\Figures\Correlation Between Numerical Features", dpi=300, bbox_inches="tight")

plt.show()
```

```python
orders_by_hour = df['order_hour'].value_counts().sort_index()

plt.figure(figsize=(10,6))

sns.lineplot(x=orders_by_hour.index,
             y=orders_by_hour.values,
             marker='o')

plt.title("Orders by Hour of Day", fontsize=14, fontweight="bold")
plt.xlabel("Hour")
plt.ylabel("Number of Orders")
plt.savefig(r"C:\Users\sainudheen\OneDrive - KEO International Consultants\Desktop\EDA_Project\Reports\Figures\Orders by Hour of Day", dpi=300, bbox_inches="tight")

plt.show()
```

```python
channel_spend = df.groupby("order_channel")["total_spend"].mean().sort_values()

plt.figure(figsize=(10,6))

channel_spend.plot(kind="bar", color="skyblue")

plt.title("Average Spending by Order Channel", fontsize=14, fontweight="bold")
plt.xlabel("Order Channel")
plt.ylabel("Average Spend")
plt.savefig(r"C:\Users\sainudheen\OneDrive - KEO International Consultants\Desktop\EDA_Project\Reports\Figures\Average Spending by Order Channel", dpi=300, bbox_inches="tight")

plt.show()
```

```python
plt.figure(figsize=(10,6))

sns.barplot(x='is_rewards_member',hue='is_rewards_member', y='total_spend', data=df, palette="pastel")

plt.title("Rewards Members vs Non-Members Spending", fontsize=14, fontweight="bold")
plt.xlabel("Rewards Member")
plt.ylabel("Average Spend")
plt.savefig(r"C:\Users\sainudheen\OneDrive - KEO International Consultants\Desktop\EDA_Project\Reports\Figures\Rewards Members vs Non-Members Spending", dpi=300, bbox_inches="tight")

plt.show()
```

```python
plt.figure(figsize=(10,6))

sns.scatterplot(x='cart_size',
                y='total_spend',
                data=df,
                alpha=0.4)

plt.title("Cart Size vs Total Spend", fontsize=14, fontweight="bold")
plt.xlabel("Cart Size")
plt.ylabel("Total Spend")
plt.savefig(r"C:\Users\sainudheen\OneDrive - KEO International Consultants\Desktop\EDA_Project\Reports\Figures\Cart Size vs Total Spend", dpi=300, bbox_inches="tight")

plt.show()
```
