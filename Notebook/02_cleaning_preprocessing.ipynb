## Data Cleaning

```python
import pandas as pd
import numpy as np

df = pd.read_csv(r"C:\Users\sainudheen\Downloads\starbucks_customer_ordering_patterns.csv")
df.head()
```

## Fix Data Type

```python
df.dtypes
```

```python
df['order_date'] = pd.to_datetime(df['order_date'], format='%Y-%m-%d')

df['order_time'] = pd.to_datetime(df['order_time'], format='%H:%M')
```

```python
df.dtypes
```

## Standardize Categorical Variables

```python
df['customer_gender'] = df['customer_gender'].str.lower().str.strip()

df['order_channel'] = df['order_channel'].str.lower().str.strip()

df['drink_category'] = df['drink_category'].str.lower().str.strip()

df['region'] = df['region'].str.lower().str.strip()
```

## Find Duplicates

```python
df.duplicated().sum()
```

## Find the Null Values

```python
df.isnull().sum()
```

### Outlier Detection (IQR Method)

```python
Q1 = df['total_spend'].quantile(0.25)
Q3 = df['total_spend'].quantile(0.75)

IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

df = df[(df['total_spend'] >= lower) & (df['total_spend'] <= upper)]
```

```python
outliers_check = df[(df['total_spend'] < lower) | (df['total_spend'] > upper)]

outliers_check.shape
```
