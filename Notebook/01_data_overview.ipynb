## Business Context

**Domain:** Retail Analytics / Customer Behavior  

**Dataset:** Starbucks customer ordering patterns (~100,000 records)

**Objective:**  
The objective of this project is to analyze Starbucks customer ordering data to identify patterns in purchasing behavior, customer demographics, product preferences, and spending habits.


## Dataset Overview

The dataset contains customer transaction data from Starbucks stores. Each record represents an individual order placed by a customer.

The dataset includes information such as:

- Customer demographics (age group, gender, rewards membership)
- Order details (order channel, order time, day of week)
- Product information (drink category and customizations)
- Spending metrics (total spend per order)
- Operational metrics (fulfillment time)
- Customer satisfaction ratings

These variables help analyze customer purchasing behavior and service performance.


## Problem Statement

Understanding customer behavior is important for improving service quality and increasing sales.

Customer purchasing patterns may vary depending on factors such as:

- Customer demographics
- Order channels (mobile, drive-thru, in-store)
- Product preferences
- Order timing
- Rewards membership

This project aims to explore Starbucks ordering data to identify patterns in customer behavior and spending.
