# HYPOTHESIS TESTING

**Hypothesis testing is used to determine whether observed patterns in the dataset are statistically significant or occurred by chance.
In this analysis, statistical tests are performed to evaluate relationships between customer behavior and spending patterns.**


```python
import pandas as pd
from scipy import stats
```

```python
df = pd.read_csv(r"C:\Users\sainudheen\OneDrive - KEO International Consultants\Desktop\EDA_Project\Data\cleaned_data_day2.csv")

df.head()
```

## Independent T-Test

**The independent t-test was conducted to determine whether there is a statistically significant difference in the average spending between rewards members and non-members.**

**H₀: There is no significant difference in the mean total spending between rewards members and non-members.**

**H₁: There is a significant difference in the mean total spending between rewards members and non-members.**

```python
members = df[df["is_rewards_member"] == True]["total_spend"]
non_members = df[df["is_rewards_member"] == False]["total_spend"]

t_stat, p_value = stats.ttest_ind(members, non_members)

print("T-statistic:", t_stat)
print("P-value:", p_value)

if p_value < 0.05:
    print("Reject H0 → Rewards membership affects spending.")
else:
    print("Fail to reject H0 → No significant difference.")
```



**• Since the p-value is less than 0.05, we reject the null hypothesis (H₀).**

**• This indicates that there is a statistically significant difference in spending between rewards members and non-members.**

**• Rewards members tend to spend more on average compared to non-members.**

**• This suggests that the loyalty program has a positive impact on customer spending behavior.**

## ANOVA Test

**The ANOVA (Analysis of Variance) test was conducted to determine whether there is a statistically significant difference in the average spending across different customer age groups.**

**H₀: There is no significant difference in the mean total spending across different customer age groups.**

**H₁: At least one customer age group has a significantly different mean spending compared to the others.**

```python

groups = [group["total_spend"].values for name, group in df.groupby("customer_age_group")]

f_stat, p_value = stats.f_oneway(*groups)

print("F-statistic:", f_stat)
print("P-value:", p_value)

if p_value < 0.05:
    print("Reject H0 → Spending differs across age groups.")
else:
    print("Fail to reject H0.")
```

**• Since the p-value is less than 0.05, we reject the null hypothesis (H₀).**

**• This indicates that average spending varies significantly across different customer age groups.**

**• The results suggest that age demographics play an important role in influencing customer spending behavior.**

**• Certain age groups contribute more to total revenue compared to others.**

## Chi-Square Test

**The Chi-Square test was conducted to determine whether there is a statistically significant relationship between region and order channel used by customers.**

**H₀: There is no significant relationship between region and order channel.**

**H₁: There is a significant relationship between region and order channel.**

```python


contingency = pd.crosstab(df["region"], df["order_channel"])

chi2, p, dof, expected = stats.chi2_contingency(contingency)

print("Chi-square:", chi2)
print("P-value:", p)
if p < 0.05:
    print("Reject H0 → Region influences order channel.")
else:
    print("Fail to reject H0.")
```

**• Since the p-value is greater than 0.05, we fail to reject the null hypothesis (H₀).**

**• This indicates that there is no statistically significant relationship between region and order channel.**

**• Customers across different regions tend to use similar ordering methods, such as mobile app, drive-thru, or in-store ordering.**

**• Regional location does not appear to strongly influence how customers place their orders.**

## Feature Engineering

Feature engineering involves creating new variables from existing data to enhance analytical insights and improve model performance.

New features can reveal hidden patterns in customer behavior and improve the overall quality of analysis.

```python

df["order_hour"] = pd.to_datetime(df["order_time"], format="%H:%M:%S").dt.hour

df.head()
```

```python
df["spend_category"] = pd.cut(
    df["total_spend"],
    bins=[0,10,20,30],
    labels=["Low","Medium","High"]
)

df["large_cart"] = df["cart_size"] > 5

df["morning_order"] = df["order_hour"] < 12

df["high_customization"] = df["num_customizations"] > 2

df["spend_per_item"] = df["total_spend"] / df["cart_size"]

df.head()
```
